# Phase 3: Exploratory Data Analysis (EDA)

In this phase, we dive into the cleaned dataset to uncover patterns, trends, and relationships. EDA is essential for transforming raw numbers into business stories, helping stakeholders understand performance and customer behavior.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visual style
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

## 1. Loading Cleaned Data
We load the processed data and ensure dates are correctly parsed for time-series analysis.

```python
df = pd.read_csv('../data/processed/cleaned_data.csv', parse_dates=['Purchase Date'])
df.head()
```

## 2. Revenue Trend Analysis
Understanding how revenue fluctuates over time is crucial for identifying seasonal peaks and potential sales slumps.

```python
# Grouping by Month to see broader trends
monthly_revenue = df.resample('M', on='Purchase Date')['Calculated_Revenue'].sum()

plt.figure(figsize=(14, 6))
plt.plot(monthly_revenue.index, monthly_revenue.values, marker='o', linestyle='-', color='teal')
plt.title('Monthly Revenue Trend', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Revenue ($)', fontsize=12)
plt.grid(True)
plt.show()
```

### 💡 Business Insights - Revenue Trends
- **Seasonality:** Identify any significant spikes (e.g., holiday seasons) that suggest a need for increased inventory or marketing spend.
- **Growth Trajectory:** A steady upward or downward slope indicates the business's long-term health and market penetration.

## 3. Top Product Categories by Revenue
Focusing on the highest-grossing categories helps in optimizing product mix and marketing focus.

```python
top_categories = df.groupby('Product Category')['Calculated_Revenue'].sum().sort_values(ascending=False)

sns.barplot(x=top_categories.values, y=top_categories.index, palette='viridis')
plt.title('Top Product Categories by Revenue', fontsize=16)
plt.xlabel('Total Revenue ($)', fontsize=12)
plt.ylabel('Category', fontsize=12)
plt.show()
```

### 💡 Business Insights - Product Performance
- **Star Categories:** Identify the 'hero' categories that drive the majority of revenue. These should be the focus of loyalty programs.
- **Underperformers:** Categories with low revenue but high volume might indicate pricing issues, while low volume/low revenue items might be candidates for liquidation.

## 4. Customer Spending Analysis
Analyzing the distribution of customer spending helps in segmenting high-value customers from occasional shoppers.

```python
customer_spending = df.groupby('Customer ID')['Calculated_Revenue'].sum()

sns.histplot(customer_spending, bins=30, kde=True, color='purple')
plt.title('Distribution of Customer Total Spending', fontsize=16)
plt.xlabel('Total Spending ($)', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)
plt.show()
```

### 💡 Business Insights - Customer Behavior
- **Spending Concentration:** If the distribution is heavily skewed to the left, it indicates a large base of low-value customers. 
- **High-Value Segment:** A long tail to the right represents your VIP customers who contribute disproportionately to the bottom line.

## 5. Payment Method Analysis
Understanding how customers prefer to pay can influence partnerships with financial providers and checkout optimizations.

```python
payment_perf = df.groupby('Payment Method')['Calculated_Revenue'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=payment_perf.index, y=payment_perf.values, palette='magma')
plt.title('Revenue Contribution by Payment Method', fontsize=16)
plt.ylabel('Total Revenue ($)', fontsize=12)
plt.show()
```

### 💡 Business Insights - Payment Preferences
- **Checkout Optimization:** High usage of a specific method (e.g., Credit Card) suggests that the checkout process must be ultra-seamless for that gateway.
- **Promotion Strategy:** Low usage of certain methods might indicate an opportunity for 'cashback' or discount promotions to diversify payment options.

## 6. Correlation Heatmap
Identifying how variables like Quantity and Price influence Revenue.

```python
corr_matrix = df[['Quantity', 'Product Price', 'Calculated_Revenue']].corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap: Sales Metrics', fontsize=16)
plt.show()
```

### 💡 Business Insights - Correlations
- **Drivers of Revenue:** A high correlation between `Quantity` and `Revenue` suggests volume is the primary driver. If `Price` has a stronger correlation, the business relies more on premium product sales.

## 7. Outlier Visualization (Post-Cleaning)
Verifying that our statistical filtering has successfully contained the variance in our numerical data.

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(y=df['Product Price'], color='skyblue')
plt.title('Boxplot of Product Price')

plt.subplot(1, 2, 2)
sns.boxplot(y=df['Quantity'], color='salmon')
plt.title('Boxplot of Quantity')

plt.tight_layout()
plt.show()
```

### 💡 Business Insights - Data Quality
- **Consistency:** The absence of extreme 'whiskers' in these plots confirms that the data is now stable and reliable for modeling and forecasting.
