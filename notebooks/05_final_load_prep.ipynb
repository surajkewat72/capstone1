# 05 - Final Data Load and KPI Preparation
In this notebook, we calculate the primary KPIs required for the Tableau dashboard and prepare the final summarized datasets.

```python
import pandas as pd

# Load the cleaned dataset
df = pd.read_csv('../data/processed/cleaned_data.csv')
df['Purchase Date'] = pd.to_datetime(df['Purchase Date'])
df.head()
```

### 1. Compute KPIs

```python
# Total Revenue
total_revenue = df['Calculated_Revenue'].sum()
print(f'Total Revenue: ${total_revenue:,.2f}')

# Average Order Value (AOV)
aov = df['Calculated_Revenue'].mean()
print(f'Average Order Value (AOV): ${aov:,.2f}')
```

```python
# Category Penetration (Revenue by Category)
category_revenue = df.groupby('Product Category')['Calculated_Revenue'].sum().reset_index()
category_revenue['Penetration %'] = (category_revenue['Calculated_Revenue'] / total_revenue) * 100
category_revenue = category_revenue.sort_values(by='Calculated_Revenue', ascending=False)
display(category_revenue)
```

```python
# Customer Lifetime Value (LTV) Proxy - Total spending per unique Customer ID
customer_ltv = df.groupby('Customer ID')['Calculated_Revenue'].sum().reset_index()
customer_ltv.rename(columns={'Calculated_Revenue': 'LTV'}, inplace=True)
customer_ltv = customer_ltv.sort_values(by='LTV', ascending=False)
display(customer_ltv.head(10))
```

### 2. Export Summarized Data for Tableau

```python
# Export KPIs to a summary CSV
category_revenue.to_csv('../data/processed/category_kpi_summary.csv', index=False)
customer_ltv.to_csv('../data/processed/customer_ltv_summary.csv', index=False)
print('KPI summaries exported successfully to data/processed/')
```
