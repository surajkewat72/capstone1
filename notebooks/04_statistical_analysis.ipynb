# Phase 4: Statistical Analysis & Hypothesis Testing

In this phase, we move beyond visual observations to statistical validation. We will use mathematical tests to determine if the patterns we saw in EDA are statistically significant or just due to random chance. This adds a layer of scientific rigor to our capstone project.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm

# Set visual style
sns.set_theme(style="white")
plt.rcParams['figure.figsize'] = (10, 6)
```

## 1. Correlation Analysis
We calculate the Pearson Correlation coefficient to understand the strength and direction of the relationship between our key numerical variables.

```python
df = pd.read_csv('../data/processed/cleaned_data.csv')

columns_to_correlate = ['Quantity', 'Product Price', 'Calculated_Revenue']
corr_matrix = df[columns_to_correlate].corr()

sns.heatmap(corr_matrix, annot=True, cmap='RdBu_r', center=0)
plt.title('Correlation Matrix of Sales Metrics')
plt.show()
```

### 🔍 Interpretation
- **Positive Correlation:** A value close to +1 suggests that as one variable increases, the other also increases.
- **Business Context:** If `Calculated_Revenue` is more highly correlated with `Quantity` than `Price`, it indicates that the business is volume-driven rather than margin-driven.

## 2. Hypothesis Testing: Gender-Based Revenue Differences
We want to know if there is a statistically significant difference in the average spending between **Male** and **Female** customers. This helps in targeted marketing.

### Hypotheses:
- **Null Hypothesis (H0):** There is no significant difference in the mean revenue between Male and Female customers.
- **Alternative Hypothesis (H1):** There is a significant difference in the mean revenue between Male and Female customers.

```python
# Splitting the data into two groups
male_revenue = df[df['Gender'] == 'male']['Calculated_Revenue']
female_revenue = df[df['Gender'] == 'female']['Calculated_Revenue']

# Performing the Independent T-Test
t_stat, p_value = stats.ttest_ind(male_revenue, female_revenue, equal_var=False)

print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Conclusion based on alpha = 0.05
if p_value < 0.05:
    print("Conclusion: Reject the Null Hypothesis (H0). There is a significant difference.")
else:
    print("Conclusion: Fail to Reject the Null Hypothesis (H0). No significant difference found.")
```

### 🔍 Business Interpretation
- **If P < 0.05:** The difference in spending is real and not due to chance. Marketing should create gender-specific campaigns to leverage this behavior.
- **If P > 0.05:** Both genders spend roughly the same amount. The business can use a gender-neutral marketing strategy to save on specialized ad costs.

## 3. Simple Linear Regression: Quantity vs Revenue
We model the relationship between `Quantity` and `Calculated_Revenue` to see how much of the revenue variance can be explained by the number of units sold.

```python
X = df['Quantity']
y = df['Calculated_Revenue']

# Adding a constant for the intercept
X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
print(model.summary())
```

### 🔍 Interpretation
- **R-squared:** This value (0 to 1) tells us what percentage of revenue variation is explained by quantity. A high R-squared means quantity is a very strong predictor of revenue.
- **Coefficient:** For every 1 unit increase in `Quantity`, the revenue is expected to increase by the coefficient amount.
